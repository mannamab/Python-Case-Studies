```python
#6.4
q64_series = pd.Series(data=['A', 'B', 'C', 'AabX', 'BacX','', np.nan, 'CABA', 'dog', 'cat'])
def replace_X_dog(var):
    if type(var) == str:
        if 'X' in var:
            return var.replace('X', 'XX-XX')
        if 'dog' in var:
            return var.replace('dog', 'XX-XX')
    return var
q64_series.apply(replace_X_dog)
```
